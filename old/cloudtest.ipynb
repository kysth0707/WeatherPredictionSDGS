```python
def getStretchedImg(source):
	resizedImg = cv.resize(source, (256, 256), interpolation=cv.INTER_AREA)
	grayScaledImage = cv.cvtColor(resizedImg, cv.COLOR_BGR2GRAY)

	minVal = min(grayScaledImage.flatten())
	maxVal = max(grayScaledImage.flatten())

	stretchedImg = (grayScaledImage - minVal) / (maxVal - minVal) * 255
	stretchedImg = stretchedImg.astype(np.uint8)
	return stretchedImg
```

```python
plt.imshow(getStretchedImg(cv.imread('testclouds/1.jpg')), cmap='gray')
```

```python
plt.imshow(getStretchedImg(cv.imread('testclouds/3.jpg')), cmap='gray')
```

```python
plt.imshow(getStretchedImg(cv.imread('testclouds/8.jpg')), cmap='gray')
```
